```python
#The pourpose of this code is to transcribe DNA into RNA

#Solution using a loop
with open("/home/lm463/jupyter_notebooks/Rosalind/rosalind_rna.txt", "r") as myfile:
    DNA_String=myfile.read().rstrip('\n')

RNA = ""
Test= "GATGGAACTTGACTACGTAAATT"
def DNAtoRNA(sequence):
    RNA = ""
    for nuc in sequence:
        if nuc == "A":
            RNA = RNA + "A"
        elif nuc == "C":
            RNA = RNA + "C"
        elif nuc == "G":
            RNA = RNA + "G"
        elif nuc == "T":
            RNA = RNA + "U"
    return RNA
DNAtoRNA(DNA_String)
```
